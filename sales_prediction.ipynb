```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.model_selection import train_test_split
```

Read Data

```python
sales_data=pd.read_csv("advertising.csv")
sales_data
```

check for nulls

```python
sales_data.isnull().sum()
```

No nulls present

Split data into test and train

```python
x=sales_data.drop(columns='Sales')
y=sales_data[['Sales']]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

Create Model

```python
model1=LinearRegression()
model2=RandomForestRegressor()
```

Fit model

```python
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
```

Predict

```python
lin_pred=model1.predict(x_test)
ran_pred=model2.predict(x_test)
```

Accuracy Score

```python
print("Linear Regression: ")
print("R2 score: ",r2_score(y_test,lin_pred))
print("Mean Absolute Error: ",mean_absolute_error(y_test,lin_pred))
print("Mean Squared Error: ",mean_squared_error(y_test,lin_pred))

print("Random Forest Regression: ")
print("R2 score: ",r2_score(y_test,ran_pred))
print("Mean Absolute Error: ",mean_absolute_error(y_test,ran_pred))
print("Mean Squared Error: ",mean_squared_error(y_test,ran_pred))
```
